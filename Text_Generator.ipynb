```python
import gradio as gr
from transformers import pipeline

# Pretrained model pipeline (text generation)
generator = pipeline("text-generation", model="gpt2")

def generate_text(prompt):
    result = generator(prompt, max_length=60, num_return_sequences=1)
    return result[0]['generated_text']

# Gradio Interface
iface = gr.Interface(
    fn=generate_text,
    inputs=gr.Textbox(lines=2, placeholder="Enter a prompt here..."),
    outputs="text",
    title="Text Generator",
    description="A simple text generation app using GPT-2 on Hugging Face."
)

if __name__ == "__main__":
    iface.launch()
```
